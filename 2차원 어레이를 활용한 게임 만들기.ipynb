iloc. loc 으로 정보 읽어내고 리스트와 비교해서 참 거짓 만들기

```python
def tikO():
    a=int(input("O의 행을 입력하시오"))
    b=int(input("O의 열을 입력하시오"))

    if '-' in grid[a][b]:
        grid[a][b]='O'

    else:
        print('이미 채워진 칸 입니다')
        print('다시 입력 하십시오')  
        tikO()
    
    
def tikX():
    a=int(input("X의 행을 입력하시오"))
    b=int(input("X의 열을 입력하시오"))

    if '-' in grid[a][b]:
        grid[a][b]='X'
    else:
        print('이미 채워진 칸 입니다')
        print('다시 입력 하십시오')  
        tikX()       
```

```python
def check(grid):
    digri=grid.diagonal() #대각선을 리스트 하나로
    
    a,b,c='','',''
    a=digri[0]
    b=digri[1]
    c=digri[2]
    if a==b==c and a=='X':
        print(f'{a}의 승리')
        return 0

    if a==b==c and a=='O':
        print(f'{a}의 승리')
        return 0

        
    
def checkrow(grid):
    a,b,c='','',''
    for i in range(3):
        a=grid[i][0]
        b=grid[i][1]
        c=grid[i][2]
        if a==b==c and a=='X':
            print(f'{a}의 승리')
            return 0

        if a==b==c and a=='O':
            print(f'{a}의 승리')
            return 0

            
def checkcol(grid):
    a,b,c='','',''
    grat=grid.transpose()
    a,b,c='','',''
    for i in range(3):
        a=grat[i][0]
        b=grat[i][1]
        c=grat[i][2]
        if a==b==c and a=='X':
            print(f'{a}의 승리')
            return 0

        if a==b==c and a=='O':
            print(f'{a}의 승리')
            return 0
```

```python
import numpy as np 
grid = np.array([["-", "O", "-"],
        ["-", "O", "-"],
        ["-", "O", "-"]])
win=1  
# grid= [['-' for i in range(3)]for j in range(3)] # 판 생성
# grid=np.array(grid)

while(1):

    win1=check(grid)
    if win1 == 0:
        break
    win2=checkrow(grid)
    if win2 == 0:
        break
    win3=checkcol(grid)
    if win3 == 0:
        break 


    tikO() # O 값받기

    print(*grid,sep='\n')
    print('')

    tikX() # X 값 받기
    
    win1=check(grid)
    if win1 == 0:
        break
    win2=checkrow(grid)
    if win2 == 0:
        break
    win3=checkcol(grid)
    if win3 == 0:
        break 

    print(*grid,sep='\n')
    print('')

    win1=check(grid)
    if win1 == 0:
        break
    win2=checkrow(grid)
    if win2 == 0:
        break
    win3=checkcol(grid)
    if win3 == 0:
        break 

    if win == 0:
        break
        

        #승리



```
